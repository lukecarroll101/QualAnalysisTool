```python
focus = "Environment_Text"
df = df[[focus]]
df.head()
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")
print("Natural Language Processor loaded")
```

```python
continuous_string = df[focus].str.cat(sep='" "')
continuous_string = '"' + continuous_string + '"'
```

```python
doc = nlp(continuous_string)
# filtered_tokens = [token.text for token in doc if not token.is_stop and not token.is_punct]
# print(filtered_tokens)

noun_phrases = [chunk.text for chunk in doc.noun_chunks]
keywords = [token.lemma_ for token in doc if token.pos_ in ("NOUN","VERB", "ADJ", "ADV")]
print(keywords)
```

```python
from spacy.matcher import Matcher

matcher = Matcher(nlp.vocab)
noun_verb_pattern = [{'POS':'NOUN'}, {'POS':'VERB'}]
noun_verb_noun_pattern = [{'POS':'NOUN'}, {'POS':'VERB'}, {'POS':'NOUN'}]
noun_verb_adj_pattern = [{'POS':'NOUN'}, {'POS':'VERB'}, {'POS':'ADJ'}]
adj_noun_pattern = [{'POS':'ADJ'}, {'POS':'NOUN'}]
verb_adj_noun_pattern = [{'POS': 'VERB'}, {'POS': 'ADJ'}, {'POS': 'NOUN'}]
noun_clause_pattern = [{'POS': 'NOUN'}, {'ENT_TYPE': 'RELCL'}]
verb_object_pattern = [{'POS': 'VERB'}, {'DEP': 'dobj'}]
noun_prep_noun_pattern = [{'POS': 'NOUN'}, {'POS': 'ADP'}, {'POS': 'NOUN'}]
verb_noun_noun_pattern = [{'POS':'VERB'}, {'POS':'NOUN'}, {'POS':'NOUN'}]
adj_noun_verb_pattern = [{'POS':'ADJ'}, {'POS':'NOUN'}, {'POS':'VERB'}]

matcher.add("NOUN_VERB", [noun_verb_pattern])
matcher.add("NOUN_VERB_NOUN", [noun_verb_noun_pattern])
matcher.add("NOUN_VERB_ADJ", [noun_verb_adj_pattern])
matcher.add("ADJ_NOUN", [adj_noun_pattern])
matcher.add("VERB_ADJ_NOUN", [verb_adj_noun_pattern])
matcher.add("NOUN_CLAUSE", [noun_clause_pattern])
matcher.add("VERB_OBJECT", [verb_object_pattern])
matcher.add("NOUN_PREP_NOUN", [noun_prep_noun_pattern])
matcher.add("VERB_NOUN_NOUN", [verb_noun_noun_pattern])
matcher.add("ADJ_NOUN_VERB", [adj_noun_verb_pattern])

matches = matcher(doc)
collocations = [doc[start:end].text for match_id, start, end in matches]
print("Collocations:", collocations)
lemmatized_collocations = []
for collocation in collocations:
    tokens = nlp(collocation)
    lemmas = " ".join([token.lemma_ for token in tokens])
    lemmatized_collocations.append(lemmas)
print("Lemmatised collocations:", lemmatized_collocations)
```

```python
from collections import Counter
keywords = [word for word in keywords if len(word) > 2 and word not in ["get", "have", "come"]]
keyword_freq = Counter(keywords)
most_common_keywords = keyword_freq.most_common(30)
print(most_common_keywords)
```

```python
# Find sentences containing 3 or more common keywords
co_occurrences = []
df2 = pd.DataFrame(columns=['sentence', 'keywords'])
df2

for i, sent in enumerate(doc.sents):
    sent_keywords = [token.lemma_ for token in sent if token.lemma_ in dict(most_common_keywords)]
    if len(sent_keywords) >= 2:
        df2.loc[i] = {'sentence': str(sent).replace('"',''), 'keywords': sent_keywords}
        co_occurrences.append((sent, sent_keywords))

df2.to_csv('keySentences.csv')
df2.head(15)
```

```python
def extract_themes(noun_phrases, most_common_keywords):
    themes = []
    for phrase in noun_phrases:
        for keyword, _ in most_common_keywords:
            if keyword in phrase:
                themes.append(phrase)
                break
    return list(set(themes))

themes = extract_themes(collocations, most_common_keywords)
print("Themes:", themes)
```
